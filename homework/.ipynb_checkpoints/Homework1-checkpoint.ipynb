## Exercise Four - Movie Lens Data

```python
import pandas as pd
import numpy as np
pd.read_table('u.user', header=None)
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = pd.read_table('u.user', sep='|', header=None, names=user_cols, index_col='user_id', dtype={'zip_code':str})

# for each occupation in 'users', count the number of occurrences
users.occupation.value_counts()
```

```python
# for each occupation, calculate the mean age
occup = users.groupby('occupation')
occup.age.mean()
```

```python
# for each occupation, calculate the minimum and maximum ages
occup.age.min(), occup.age.max()
```

```python
# for each combination of occupation and gender, calculate the mean age
occupage = users.groupby(['occupation','gender'])
occupage.age.mean()
```

```python
# randomly sample a DataFrame
rand_users = pd.Series(np.random.randn(1000))
rand_users.head()

users.take(np.random.permutation(len(users))[:5])
```

```python
# detect duplicate users
#users.groupby(['age','gender','occupation','zip_code']).value_counts()
users.duplicated()
duplicate = users.groupby(['age','gender','occupation','zip_code']).size().reset_index().rename(columns={0:'count'})
duplicate
```
