```python
#using Melbourne dataset trying to predict Median HousePrice based on the historical dataset (Deep Neural Network)
"""Example of DNNRegressor for Housing dataset."""

from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np
from sklearn import datasets
from sklearn import metrics
from sklearn import model_selection
from sklearn import preprocessing

import tensorflow as tf
import pandas as pd
```

```python
 # Load dataset
melbourne_df = pd.read_csv('Melbourne_housing_extra_data.csv')
melbourne_df.head()
```

```python
#Feature Engineering

import pandas_profiling
pandas_profiling.ProfileReport(melbourne_df)
```

```python
#change Type to Numeric
melbourne_df.Type.value_counts()
melbourne_df['Type'] = melbourne_df.Type.map({'h':0, 'u':1, 't':2})

# fill in missing values for Features with Mean for each respected Feature
melbourne_df.Bathroom.fillna(melbourne_df.Bathroom.mean(), inplace=True)
melbourne_df.BuildingArea.fillna(melbourne_df.BuildingArea.mean(), inplace=True)
melbourne_df.Car.fillna(melbourne_df.Car.mean(), inplace=True)
melbourne_df.Landsize.fillna(melbourne_df.Landsize.mean(), inplace=True)
melbourne_df.Lattitude.fillna(melbourne_df.Lattitude.mean(), inplace=True)
melbourne_df.Longtitude.fillna(melbourne_df.Longtitude.mean(), inplace=True)
melbourne_df.Price.fillna(melbourne_df.Price.mean(), inplace=True)
melbourne_df.YearBuilt.fillna(melbourne_df.YearBuilt.mean(), inplace=True)
```

```python
# Split dataset into train / test with 20% value for Test data

feature_cols = ['Rooms', 'Type', 'Distance','Bathroom', 'Car', 'Landsize',
               'BuildingArea', 'YearBuilt']
x = melbourne_df[feature_cols]
y = melbourne_df.Price

x_train, x_test, y_train, y_test = model_selection.train_test_split(
x, y, test_size=0.2, random_state=42)
```

```python
 # Scale data (training set) to 0 mean and unit standard deviation.
scaler = preprocessing.StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
#Build 2 layer fully connected DNN with 10, 10 units respectively.
#Instantiate a DNNRegressor for the neural network regression model. 
#We need to provide two arguments here: 
#Hidden_units, a hyperparameter specifying the number of nodes in each hidden layer (here, two hidden layers with 10 nodes each), and feature_columns
feature_columns = [
      tf.feature_column.numeric_column('x', shape=np.array(x_train).shape[1:])]
regressor = tf.contrib.learn.DNNRegressor(
      feature_columns=feature_columns, hidden_units=[10, 10])
```

```python
# Train the dataset
train_input_fn = tf.estimator.inputs.numpy_input_fn(
x={'x': x_train}, y=y_train, batch_size=1, num_epochs=None, shuffle=True)
regressor.fit(input_fn=train_input_fn, steps=2000)
```

```python
# Predict.
x_transformed = scaler.transform(x_test)
test_input_fn = tf.estimator.inputs.numpy_input_fn(
x={'x': x_transformed}, y=y_test, num_epochs=1, shuffle=False)
predictions = regressor.predict_scores(input_fn=test_input_fn)
y_predicted = np.array(list(p for p in predictions))
y_predicted = y_predicted.reshape(np.array(y_test).shape)
y_predicted
```

```python
# Score with sklearn.
score_sklearn = metrics.mean_squared_error(y_predicted, y_test)
print('MSE (sklearn): {0:f}'.format(score_sklearn))

# Score with tensorflow.
scores = regressor.evaluate(input_fn=test_input_fn)
loss_score = scores["loss"]
print('MSE (tensorflow): {0:f}'.format(loss_score))
```
